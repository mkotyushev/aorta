```python
import plotly.graph_objects as go
import pandas as pd

df = pd.read_csv('/workspace/data/results-imagenet-real.csv')
df['param_count'] = df['param_count'].str.replace(',', '').astype(float)
df['top1'] = df['top1'].astype(float)

layout = dict(plot_bgcolor='white',
              margin=dict(t=20, l=20, r=20, b=20),
              xaxis=dict(title='Parameter Count',
                         range=[0, 3.2],
                         linecolor='#d9d9d9',
                         showgrid=False,
                         mirror=True),
              yaxis=dict(title='Top 1 Accuracy',
                         range=[64, 92],
                         linecolor='#d9d9d9',
                         showgrid=False,
                         mirror=True))

data = go.Scatter(x=df['param_count'],
                  y=df['top1'],
                  text=df['model'],
                  textposition='top right',
                  textfont=dict(color='#E58606'),
                  mode='markers',
                  marker=dict(color='#5D69B1', size=8),
                  name='citations')

fig = go.Figure(data=data, layout=layout)
fig.update_xaxes(type="log")

fig.show()
```
